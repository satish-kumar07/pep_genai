```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,accuracy_score,confusion_matrix,classification_report
```

```python
# df = load_iris()
```

```python
df = pd.read_csv('Iris.csv')
```

```python
X = df.iloc[:,1:5].values
y = df['Species']
# X = df.data
# y = df.target
```

```python
X
```

```python
n = 69
```

```python
X_train,X_test ,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=n)
```

```python
model = DecisionTreeClassifier(random_state=n)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    'max_depth':range(1,10,1),
    'min_samples_leaf':range(1,20,2),
    'min_samples_split':(2,20,2),
    'criterion':["entropy","gini"]
}

tree = DecisionTreeClassifier(random_state=n)

grid_search = GridSearchCV(estimator=tree,param_grid=param_grid,cv=5,verbose=True)

grid_search.fit(X_train,y_train)

print("Best ACC: ",grid_search.best_score_)

print(grid_search.best_estimator_)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

tree_clf = grid_search.best_estimator_

plt.figure(figsize=(20,4))

plot_tree(tree_clf,filled=True,feature_names=X,class_names=y)

plt.show()
```
