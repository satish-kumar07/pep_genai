## Decision Tree Model
* A decision tree in a supervised ML algo used for both classification[predict the categories] and Regression[Predic the continous value] task. 
    
    It mimics human decision making by splitting the data based on features value in an upside-down , flowchart-like tree structure.

### Anatomy of Decision tree

1. Root Node -> The very top Node representing the entire dataset and the first feature test.

2. Decision/internal Node -> intermediate node where futhere condition or data split occurs.

3. Branches -> Connecting lines that represent the outcome of a decision or condition rule.

4. Leaf/terminal node ->  the final endpoint that hold the prediction outcomes or close label.

### Core Limitations

1. Overfitting - high depth trees easily over-memorize training data.
