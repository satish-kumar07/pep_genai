# Clustering

- Clustering is a unsupervised learning algorithm, used to gropup similar data points together predefined labels
- The ultimate goal is to sort the data so that objects in the same group[cluster] are highly similat to each other, 
    while the objects in the group are so distinct as possible.
- Because it operates on unlabeled data , clustering is primanily used to discovers the hidden patterns, structure or relationships within complex datasets.


#### Core Approach of Clustering 

clustering methods gradually falls into two broad classifications based on how strictly points are assigned.

1.  Hard Clustering: Each data points belong to exactly one sample cluster with a clear boundary.
2. Soft Clustering: Data points can belong to multiple cluster, each associated with a probability or degreen if membership.

#### Popular Clustering Algorithm

Different Algorithm use different mathmatica, approaches to find these groups[clusters]

1. Centroid-Based Clustering:

    * Group data points around central points called centeroid
    
    * Key Algorithm K-Means, Partition a dataset into a predefined number[k] of spherical cluster by maintaing minimum data points distance, and    ther prespective cluster center.
    
    * Pros & Cons - It is incredibly fast and scalable, but it requires you to choose the value of K. beforehand and struggles with non-spherical shapes.

2. Connecting Baased Clstering[Herarichal clustering]

    * Builds a tree like Strture of clusters, Called the Dengogram, by either iteratively merging smaller cluster[agglomerative] or spliting the smaller cluster[Divisive].

    * Key - Algorithm -> Agglomertaive Clustering[bottom-Top]

    * Pros & Cons -> Highly intuitive and does not requires a predefined no. of Clusters, However it is Computationally Expensive and Slow on large Datasets.

3. Density Based Clustering

    * Locate area of high data points density and seperate them by region of low Density.

    * Key Algorithm: DB-SCAN[```Density Based Spatial Clustering of application with Noise```]

    * Pros & Cons: It automatically determines the no of cluster and outliers[Noise]. However it struggles if the dataset features have highly varing density

4. Distribution-Based Clustering

    * Assumes that the dataset is made up of mixture of Distincts underlying probability distribution.

    * Key-Algorithm: Gaussian Mixture Model[GMM] :- Which calculates the probabilty that a data points belong to a specific gaussian Distribution.

    * Pros & Cons: Excellent for Comples, Elongates or ellipsdil cluster shape, but highly reluant on Statistical assumptions.
