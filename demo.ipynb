#### Acceptance value

1. LowerBond = Q1 - 1.5*(IQR)
2. UpperBond = Q3 + 1.5*(IQR)

    IQR = Q3 - Q1


```python
Q1 = df['Area (mean)'].quantile(0.25)
Q3 = df['Area (mean)'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
LowerBond = Q1 - 1.5*(IQR)
UpperBond = Q3 + 1.5*(IQR)
```

```python
print("LowerBond: ",LowerBond)
print("UpperBond: ",UpperBond)
```

```python
outliersarea = df[(df['Area (mean)'] < LowerBond) | (df['Area (mean)'] > UpperBond)]
outliersarea
```
